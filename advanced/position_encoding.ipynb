```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

def pos_enc(seq_length, d=512, n=10000):
    p = np.zeros((seq_length, d))
    for k in range(seq_length):
        for i in np.arange(int(d/2)):
            denom = n**(2*i/d)
            p[k, 2*i] = np.sin(k/denom)
            p[k, 2*i+1] = np.cos(k/denom)
    return p
```

```python
x = np.arange(100)
p = pos_enc(seq_length=1000, d=512)
plt.imshow(p)
plt.show()
```
